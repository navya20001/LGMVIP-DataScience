# TASK NAME-IMAGE TO PENCIL SKETCH WITH PYTHON

```python
import cv2
import numpy as np
```

```python
from PIL import Image as im
```

```python
image=cv2.imread('micky.png')
```

```python
image
```

CONVERTING TO GRAY IMAGE AND DISPLAYING

```python
grey_filter=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
```

```python
grey_filter.shape
```

```python
grey=im.fromarray(grey_filter)
```

```python
grey
```

CONVERTING TO INVERT AND DISPLAYING

```python
invert=cv2.bitwise_not(grey_filter)
```

```python
invert.shape
```

```python
inv=im.fromarray(invert)
```

```python
inv
```

CONVERTING TO BLUR AND DISPLAYING THE IMAGE

```python
blur=cv2.GaussianBlur(invert,(21,21),0)
```

```python
blur.shape
```

```python
b1=im.fromarray(blur)
```

```python
b1
```

CONVERTING TO INVERTEDBLUR AND DISPLAYING 

```python
invertedblur=cv2.bitwise_not(blur)
```

```python
invertedblur.shape
```

```python
inve=im.fromarray(invertedblur)
```

```python
inve
```

#CONVERTING TO PENCIL SKETCH FORM AND DISPLAYING

```python
sketch_filter=cv2.divide(grey_filter,invertedblur,scale=256.0)
```

```python
cv2.imwrite('output.jpg',sketch_filter)
```

```python
sketch_filter.shape
```

```python
data=im.fromarray(sketch_filter)
```
